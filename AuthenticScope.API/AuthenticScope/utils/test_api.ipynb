```python
import requests

url = "https://api.elevenlabs.io/v1/voices"

headers = {
  "Accept": "application/json",
  "xi-api-key": api_key
}

response = requests.get(url, headers=headers)

print(response.text)
```

```python
response = json.loads(response.text)
print(len(response['voices']))
```

```python
labels = [response['voices'][i]['labels'] for i in range(len(response['voices']))]
```

```python
label = labels[:-1]
```

```python
labels2 = []
for i in label:
    temp={}
    if 'accent ' in i.keys() and 'description ' in i.keys():
        temp['accent']=i['accent ']
        temp['age']=i['age']
        temp['gender']=i['gender']
        temp['use case']=i['use case']
        temp['description']=i['description ']
    
    elif 'accent ' in i.keys():
        temp['accent']=i['accent ']
        temp['age']=i['age']
        temp['gender']=i['gender']
        temp['use case']=i['use case']
        temp['description']=i['description']
    
    elif 'usecase' in i.keys():
        temp['accent']=i['accent']
        temp['age']=i['age']
        temp['gender']=i['gender']
        temp['use case']=i['usecase']
        temp['description']=i['description ']

    elif 'description ' in i.keys():
        temp['accent']=i['accent']
        temp['age']=i['age']
        temp['gender']=i['gender']
        temp['use case']=i['use case']
        temp['description']=i['description ']

    else:
        temp['accent']=i['accent']
        temp['age']=i['age']
        temp['gender']=i['gender']
        temp['use case']=i['use case']
        temp['description']=i['description']
    
    labels2.append(temp)
```

```python
import pandas as pd
# for person in response['voices']:
df = pd.DataFrame({'names':[person['name'] for person in response['voices'][:-1]],\
                  'accent':[labels2[i]['accent'] for i in range(len(labels2))],\
                  'description': [labels2[i]['description'] for i in range(len(labels2))],\
                  'age':[labels2[i]['age'] for i in range(len(labels2))],\
                  'gender':[labels2[i]['gender'] for i in range(len(labels2))],\
                  'use_case':[labels2[i]['use case'] for i in range(len(labels2))],
                  'voice_id':[i['voice_id'] for i in response['voices'][:-1]]})
    
```

```python

url = "https://api.elevenlabs.io/v1/text-to-speech/OZ8cofHhtZUIYPPbURk6"

import requests

CHUNK_SIZE = 1024

headers = {
  "Accept": "audio/mpeg",
  "Content-Type": "application/json",
  "xi-api-key": api_key
}

data = {
  "text": "Hi! My name is sneha, nice to meet you harry!. what are you doing?",
  "model_id": "eleven_monolingual_v1",
  "voice_settings": {
    "stability": 0.5,
    "similarity_boost": 0.5
  }
}

response = requests.post(url, json=data, headers=headers)
with open('output2.mpeg', 'wb') as f:
    for chunk in response.iter_content(chunk_size=CHUNK_SIZE):
        if chunk:
            f.write(chunk)
```
